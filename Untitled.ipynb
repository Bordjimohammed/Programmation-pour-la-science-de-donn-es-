```python
import pandas as pd
df=pd.read_csv('Learners_Digital_clean.csv')
```

```python
df
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
apprenant_cours = df['Cours'].value_counts()
print(apprenant_cours)
plt.figure(figsize=(8, 5))
sns.barplot(x=apprenant_cours.index,y=apprenant_cours.values, palette='Set3')
plt.title("Nombres d'apprenant par cours")
plt.xlabel("Cours")
plt.ylabel("Nombre d'apprenats ")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
df['Score_Quiz_%']=df['Score_Quiz_%'].astype(float)
plt.figure(figsize=(10,6))
plt.hist(df['Score_Quiz_%'], bins=20, color='pink', alpha=0.5, edgecolor='red')
plt.title("Distribution des Scores")
plt.xlabel("Score %")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Cours', y='', hue='Score_Quiz_%', palette='Set3')
plt.title("score de chaque cours")
plt.xlabel("cours")
plt.ylabel("score")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Activités_Complétées'],df['Score_Quiz_%'])
plt.xlabel("Activités Complétées")
plt.ylabel("Score Quiz")
plt.title("Relation entre Activité et score")
plt.grid(True)
plt.show()

# === Calcul et affichage du coefficient de corrélation ===
r = df['Activités_Complétées'].corr(df['Score_Quiz_%'])
print(f"Corrélation (Activités_Complétées, Score_Quiz_%) : r = {r:.3f}")
```
